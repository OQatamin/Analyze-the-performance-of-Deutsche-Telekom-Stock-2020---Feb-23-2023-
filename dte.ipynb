```python
import pandas as pd 
import numpy as np

data = pd.read_csv('DTE.DE.csv')

data.describe()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date')
```

```python
data['Daily_Return'] = data['Adj Close'].pct_change()
data = data.dropna()
print(data)
```

```python
bull_market_threshold = 0.003
bear_market_threshold = -0.003
data['Market_Type'] = 'Neutral'
data.loc[data['Daily_Return'] > bull_market_threshold, 'Market_Type'] = 'Bull'
data.loc[data['Daily_Return'] < bear_market_threshold, 'Market_Type'] = 'Bear'
filtered_data = data[['Market_Type', 'Daily_Return']]
print(filtered_data[(data['Market_Type'] != 'Neutral') & ((data['Daily_Return'] > bull_market_threshold) | (data['Daily_Return'] < bear_market_threshold))])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
short_term_window = 20
data['Short_Term_MA'] = data['Adj Close'].rolling(window=short_term_window).mean()
long_term_window = 50
data['Long_Term_MA'] = data['Adj Close'].rolling(window=long_term_window).mean()
data['Year'] = data.index.year

for year in data['Year'].unique():
    year_data = data[data['Year'] == year]

    plt.figure(figsize=(12, 6))
    plt.plot(year_data.index, year_data['Adj Close'], label='Adj Close', color='blue')
    plt.plot(year_data.index, year_data['Short_Term_MA'], label=f'{short_term_window}-day Short Term MA', color='orange')
    plt.plot(year_data.index, year_data['Long_Term_MA'], label=f'{long_term_window}-day Long Term MA', color='green')
    plt.title(f'Deutsche Telekom Stock Price with Short-Term and Long-Term Moving Averages - {year}')
    plt.xlabel('Date')
    plt.ylabel('Adj Close Price')
    plt.legend()
    year_data['Signal'] = 0
    year_data.loc[year_data['Short_Term_MA'] > year_data['Long_Term_MA'], 'Signal'] = 1  
    year_data.loc[year_data['Short_Term_MA'] < year_data['Long_Term_MA'], 'Signal'] = -1 
    buy_signals = year_data[year_data['Signal'] == 1]
    sell_signals = year_data[year_data['Signal'] == -1]
    plt.scatter(buy_signals.index, buy_signals['Adj Close'], marker='^', color='g', label='Buy Signal')
    plt.scatter(sell_signals.index, sell_signals['Adj Close'], marker='v', color='r', label='Sell Signal')
    plt.legend()

    plt.show()
```

```python
data['std'] = data['Close'].rolling(window=short_term_window).std()
data['Upper_Band'] = data['Short_Term_MA'] + 2 * data['std']  # Adjust as needed
data['Lower_Band'] = data['Short_Term_MA'] - 2 * data['std']  # Adjust as needed
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data['Signal'] = 0  
data.loc[data['Short_Term_MA'] > data['Long_Term_MA'], 'Signal'] = 1  
data.loc[data['Short_Term_MA'] < data['Long_Term_MA'], 'Signal'] = -1  
data.loc[data['Adj Close'] < data['Upper_Band'], 'Signal'] = -1  
data.loc[data['Adj Close'] > data['Lower_Band'], 'Signal'] = 1  


grouped_data = data.groupby(data.index.year)
for year, year_data in grouped_data:
    plt.figure(figsize=(12, 6))
    plt.plot(year_data.index, year_data['Adj Close'], label='Adj Close', color='blue')
    plt.plot(year_data.index, year_data['Short_Term_MA'], label=f'{short_term_window}-day Short Term MA', color='orange')
    plt.plot(year_data.index, year_data['Long_Term_MA'], label=f'{long_term_window}-day Long Term MA', color='green')
    plt.plot(year_data.index, year_data['Upper_Band'], label='Upper Bollinger Band', color='red', linestyle='dashed')
    plt.plot(year_data.index, year_data['Lower_Band'], label='Lower Bollinger Band', color='red', linestyle='dashed')
    buy_signals = year_data[year_data['Signal'] == 1]
    sell_signals = year_data[year_data['Signal'] == -1]
    plt.scatter(buy_signals.index, buy_signals['Adj Close'], marker='^', color='g', label='Buy Signal')
    plt.scatter(sell_signals.index, sell_signals['Adj Close'], marker='v', color='r', label='Sell Signal')
    plt.legend()
    plt.title(f'Deutsche Telekom Stock Price with Technical Indicators - {year}')
    plt.xlabel('Date')
    plt.ylabel('Adj Close Price')

    plt.show()
```

```python
from prophet import Prophet

data_prophet1 = data[['Adj Close']].copy()
data_prophet1.reset_index(inplace=True)
data_prophet1.columns = ['ds', 'y']
model = Prophet(daily_seasonality=True)  
model.fit(data_prophet1)
future = model.make_future_dataframe(periods=365)  
forecast = model.predict(future)
fig = model.plot(forecast)
plt.title('Deutsche Telekom Stock Price Prediction with Prophet')
plt.xlabel('Date')
plt.ylabel('Adj Close Price')
plt.show()
```
